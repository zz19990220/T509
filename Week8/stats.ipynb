```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

stats = pd.read_csv("/content/game_result.csv", header=None)
stats
```

```python
count = [0, 0]

with open("/content/game_result.csv",'r') as data:
  for row in csv.reader(data):
    if row[1] == "Bot":
        count[0] += 1
    else:
        count[1] += 1

_, ax = plt.subplots()
ax.pie(count, explode=(0, 0), labels= ['Bot', 'Human'], autopct='%1.1f%%')
ax.set_title('Bot vs Human Winning Chance')

plt.show()
```

```python
count = {}

with open("/content/game_result.csv",'r') as data:
  for line in csv.reader(data):
    count[line[0]] = count.get(line[0],0)+1

player = ["Human player - O", "Human player - X", "Bot player - X"]
count_array = [count[player[0]],count[player[1]], count[player[2]]]

fig = plt.figure(figsize = (10, 5))
 
plt.bar(player, count_array, color ='maroon', width = 0.4)
plt.xlabel("Player")
plt.ylabel("Count of wins")
plt.title("Player Rank")
plt.show()
```
